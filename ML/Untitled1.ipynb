```python
x=df.iloc[:,1:-1].values
```

```python
y=df.iloc[:,-1].values
```

```python
y
```

```python
df.tail(10)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=45)
```

```python
#split your dataset into subsets #test size indicate 30% data will be used for testing 
# random state controlls shuffling applied to the data befre applying  split
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test) # predict methoud of KNeighborsclassifier class perform prediction on testing instance
```

```python
from sklearn.metrics import confusion_matrix ,accuracy_score
cm=confusion_matrix(y_test,y_pred) #evaluate accuracy of classification
```

```python
cm
```

```python
from sklearn.metrics import classification_report
cl_report=classification_report(y_test,y_pred)
cl_report
```

```python
print("Accuracy score for KNN:",accuracy_score(y_pred,y_test))
#a ration of correctly predicated observation to the observations
```
